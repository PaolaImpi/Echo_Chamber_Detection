##### Build Vax_Twitter Graph

```python
path = "C:/MyPath/Twitter_Vaccination_Kaggle/"
```

```python
vax_df= pd.read_csv(path+ 'vaccination.csv')
```

```python
vax_df=vax_df[['id', 'date','user_id', 'username', 'tweet', 'mentions', 'reply_to']]
vax_df
```

```python
vax_df['mentions'].replace('[]', "['self']", inplace=True)
```

```python
vax_df['mentions'] = vax_df['mentions'].apply(lambda row: literal_eval(row))
```

```python
tvax_graph = nx.Graph()

#graph: tweet username ---> mentioned users

for _, row in vax_df.iterrows():
      for mention in row['mentions']:
        if tvax_graph.has_edge(row['username'], mention):
            tvax_graph[row['username']][mention]['weight'] += 1  #increment weight if the edge already exists
    
        else: #create edge and nodes if necessary
            tvax_graph.add_edge(row['username'], mention, weight = 1)
```

```python
Gcc = sorted(nx.connected_components(tvax_graph), key=len, reverse=True)
G0 = tvax_graph.subgraph(Gcc[0]) #extract giant component
```

```python
write_gexf(G0, 'Vax_twitter_graph', path)
```

##### Preprocess tweets: in text_cleaned column we save the cleaned tweet

```python
import re
```

```python
# extract hashtag 
vax_df['hashtag'] = vax_df['tweet'].apply(lambda x: re.findall(r"#(\w+)", x))
```

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt

def clean_tweets(tweets):
    #remove twitter Return handles (RT @xxx:)
    tweets = np.vectorize(remove_pattern)(tweets, "RT @[\w]*:")

    #remove twitter handles (@xxx)
    tweets = np.vectorize(remove_pattern)(tweets, "@[\w]*")
    
    #remove #
    tweets= np.core.defchararray.replace(tweets, "#", "")

    return tweets

def clean_text(text):
    #To lowercase
    text = text.lower()
    #Remove ...
    text= text.replace('...', ' ')
    #Remove newline, return, tab, form [ \n\r\t\f]
    text = re.sub(r'(\s+)', ' ', text)
    #Removing URLs and pics
    text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text)
    text = re.sub(r'pic.twitter.com/[\w]*',"", text)
    #Strip whitespaces at the beginning and at the end of text
    text = text.strip()
    
    return text
```

```python
vax_df['text_cleaned'] = clean_tweets(vax_df['tweet'])
vax_df['text_cleaned']= vax_df['text_cleaned'].apply(clean_text)
```

#### Build Reddit graphs

```python
path = "C:/MyPath/Reddit_API/"
```

```python
def preprocess_reddit_data(dvax, tvax):
    dvax['TEXT']= dvax[["Title", "Text"]].astype(str).apply(" ".join, axis=1)
    dvax= dvax[['ID', 'Post_author', 'TEXT']]
    dvax['parent_id']= dvax['ID']
    dvax.rename(columns={'ID': 'ID', 'Post_author': 'Author', 'TEXT': 'Text', 'parent_id': 'parent_id'}, inplace=True)
    tvax= tvax.drop_duplicates()
    tvax['parent_id']= tvax['parent_id'].str[3:]
    tvax= tvax[['Comment_ID', 'Comment_Author', 'Comment_Text', 'parent_id']]
    tvax.rename(columns={'Comment_ID': 'ID', 'Comment_Author': 'Author', 'Comment_Text': 'Text', 'parent_id': 'parent_id'}, inplace=True)
    return pd.concat([dvax,tvax], ignore_index=True)
```

```python
#example for reddit_vax graph
t1= pd.read_csv(path+'1000DebateVaccines_Thread.csv')
d1= pd.read_csv(path+'1000DebateVaccines.csv')
```

```python
vax_all= preprocess_reddit_data(d1, t1)
```

```python
vax_all
```

```python
vax_all['Reply_to']= None
for _, row in vax_all.iterrows():
    row['Reply_to']= vax_all[vax_all['ID']== row['parent_id'] ]['Author'].to_list()
```

```python
vax_all.to_csv('Vax_all_replies.csv', index= False)
```

```python
Rvax_graph = nx.Graph()

for _, row in vax_all.iterrows():
      for reply in row['Reply_to']:
        if Rvax_graph.has_edge(row['Author'], reply):
            Rvax_graph[row['Author']][reply]['weight'] += 1  #increment weight if the edge already exists
    
        else: #create edge and nodes if necessary
            Rvax_graph.add_edge(row['Author'], reply, weight = 1)
```
